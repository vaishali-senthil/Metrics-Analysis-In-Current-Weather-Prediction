```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dataset =pd.read_csv('vaish.csv')
```

```python
dataset.head(20)
```

```python
X = dataset.drop(labels=['weather__icon','base','sys__id','sys__country','main__sea_level','main__grnd_level','wind__gust','sys__type','timezone','cod'],axis=1)
Y = dataset['weather__description']
```

```python
Y
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label1 = LabelEncoder()
X['weather__main'] = label1.fit_transform(X['weather__main'])
X['weather__description'] = label1.fit_transform(X['weather__description'])
X['name'] = label1.fit_transform(X['name'])
```

```python
X
```

```python
X = pd.get_dummies(X, drop_first=True, columns=['weather__main'])
X.head()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 0)
```

```python
scaler = StandardScaler()
X_train =scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
#Y_train =scaler.fit_transform(Y_train.reshape(-1, 1))
#Y_test = label1.transform(Y_test)
```

```python
X_train
```

```python
model = Sequential()
model.add(Dense(X.shape[1],activation='relu',input_dim=X.shape[1]))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
X.shape[1]
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
Y_train = label1.fit_transform(Y_train)
Y_train
```

```python
model.fit(X_train,Y_train,batch_size=10, epochs=10, verbose=1)
```

```python
Y_train =scaler.fit_transform(Y_train.reshape(-1, 1))
Y_test = label1.transform(Y_test)
```

```python
model.evaluate(X_test, Y_test)
```
